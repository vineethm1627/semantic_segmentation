1) Image classification --> an entire image is classified. <br>
2) Object Detection --> detects the location of objects present in the image <br>
3) Some Computer Vision Problems require deeper understanding of the contents in the image. <br>

Classification and Object Detection may not be suitable to solve such problems. <br>

Every image is made depicted as a 2d matrix, each representing a pixel value of the digital image. <br>

Image segmentation is the task of pixel-wise classification of the image and not just the image as a whole. <br>

A Machine is able to draw useful insights more effectively by dividing the image into different segments according to the classes assigned to each of the pixel values present in the image. <br>

There are problems that require detailed information about the objects present in an image, something which is more insightful that classifying the entire image or provision of bounding boxes. <br><br>
One of the major application of <b>Image Segmentation</b> is <b>Autonomous Driving Cars for efficient understanding of the road scenes </b>

<br>
There are 2 major types of Image Segmentation: <br>
1) Semantic Segmentation: Different instances of the same object are segmented with the same color map. <br>

2) Instance Segmentation: Different instances of the same object are segmented with different color maps.

# Semantic Segmentation

```python
# importing the libraries
import pixellib
import matplotlib.pyplot as plt
from pixellib.semantic import semantic_segmentation
```

```python
# instantiating the semantic segmentation class
segment_image = semantic_segmentation()
```

 ## link to the deeplabv3+ model weights
[view](https://github.com/ayoolaolafenwa/PixelLib/releases/download/1.1/deeplabv3_xception_tf_dim_ordering_tf_kernels.h5)

```python
# xception model colormaps
pic = plt.imread("xception_model_colormap.png")
plt.figure(figsize = (10, 10))
plt.imshow(pic)
```

```python
# loading the model deeplabv3+ trained on pascal voc dataset.
segment_image.load_pascalvoc_model("deeplabv3_xception_tf_dim_ordering_tf_kernels.h5")
```

```python
# performing the segmentation on the input image
segment_image.segmentAsPascalvoc("input_images/car.jpg", output_image_name = "output_images/car.jpg")
```

```python
# performing the segmentation on the input image with overlay
segment_image.segmentAsPascalvoc("input_images/car.jpg", output_image_name = "output_images/car_overlay.jpg", overlay = True)
```

```python
img = plt.imread("input_images/car.jpg")
img_seg = plt.imread("output_images/car.jpg")
img_seg_overlay = plt.imread("output_images/car_overlay.jpg")
plt.figure(figsize = (30, 30))
plt.subplot(131)
plt.imshow(img)
plt.axis('off')
plt.xlabel("Input image")

plt.subplot(132)
plt.imshow(img_seg)
plt.axis('off')
plt.xlabel("Segmented Image")

plt.subplot(133)
plt.imshow(img_seg_overlay)
plt.axis('off')
plt.xlabel("Overlay")
```
